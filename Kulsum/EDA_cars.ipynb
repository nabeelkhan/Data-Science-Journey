## Manufacturer

```python
#Number of unique manufacturers

unique_manufacturer = df1["Make"].nunique()
print("Number of unique manufacturer :",unique_manufacturer )
```

```python
#Names of unique manufacturers

df1["Make"].unique()
```

```python
#Percentages of different manufacturer's cars

percentage = df1['Make'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
print("Percentage of different cars\n",percentage)
```

## Year

```python
#Number of different year's cars

year_count = df1['Year'].value_counts()
print("Number of cars in different years\n",year_count)
```

```python
#Percentage of differet year's cars

percentage = df1['Year'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
print("Percentage of cars in different years\n",percentage)
```

```python
oldest = df1["Year"].min()
print("Oldest car year:",int(oldest))
```

```python
latest = df1["Year"].max()
print("Latest car year:",int(latest))
```

```python
#Number of null values in each column

df1.isnull().sum()
```

## Price

```python
#sorting prices of cars in ascending order

df1 = df1.sort_values("Price")
df1
```

```python
#Lowest car amount

lowest_price_range = df1["Price"].min()
print("Lowest price range: $",lowest_price_range)
```

```python
#Highest car amount

highest_price_range = df1["Price"].max()
print("Highest price range: $",highest_price_range)
```

### Average price of each maufacturer

```python
df1["Average"]=df1.groupby(["Make"]).Price.transform('mean')
df1
```

### Percentile rank

```python
df1['Percentile Rank'] = df1.Price.rank(pct = True)
df1
```

```python
#sorting prices of cars in ascending order

df1 = df1.sort_values("Percentile Rank",ascending=False)
df1
```
